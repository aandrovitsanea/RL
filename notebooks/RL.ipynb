# Topic 2: MARKOV DECISION PROCESS AND BELLMAN PRINCIPLE OF OPTIMALITY

## Exercise: Define MDP Components

In this exercise, we'll define the basic components of an MDP: states, actions, transition probabilities, and rewards based on our previous grid world.

```python
# Define transition probabilities for each action
def get_transition_prob(state, action, grid_size):
    i, j = state
    if action == 'Up':
        next_state = (i-1 if i-1 >= 0 else i, j)
    elif action == 'Down':
        next_state = (i+1 if i+1 < grid_size else i, j)
    elif action == 'Left':
        next_state = (i, j-1 if j-1 >= 0 else j)
    elif action == 'Right':
        next_state = (i, j+1 if j+1 < grid_size else j)
    else:
        next_state = state
    
    # Simple deterministic transition
    return {next_state: 1.0}

# Test transition probabilities
current_state = (2, 2)
action = 'Up'
transition_probabilities = get_transition_prob(current_state, action, grid_size)

print(f"Transition probabilities from state {current_state} on action '{action}':")
print(transition_probabilities)
```

### Explanation:

We defined a simple deterministic transition function for each action. The function returns the next state with a transition probability of 1.0, assuming the agent remains within the grid boundaries.

## Exercise: Calculating Expected Value for an Action

Using the defined MDP components, we will calculate the expected value for a given state-action pair.

```python
# Define the reward function
def get_reward(state, grid_size):
    if state == (grid_size-1, grid_size-1):
        return 10
    return -1

# Define a function to calculate the expected value
def calculate_expected_value(state, action, grid_size, gamma, value_function):
    transition_probabilities = get_transition_prob(state, action, grid_size)
    expected_value = 0
    for next_state, prob in transition_probabilities.items():
        reward = get_reward(next_state, grid_size)
        expected_value += prob * (reward + gamma * value_function[next_state])
    return expected_value

# Initialize value function and discount factor
value_function = np.zeros((grid_size, grid_size))
gamma = 0.9

# Calculate and print expected value for a given state-action pair
current_state = (2, 2)
action = 'Right'
expected_value = calculate_expected_value(current_state, action, grid_size, gamma, value_function)
print(f"Expected value for state {current_state} and action '{action}': {expected_value}")
```

### Explanation:

We've defined a function to calculate the expected value for a given state and action based on transition probabilities, rewards, and the value function. This function helps evaluate the quality of taking a specific action from a given state.

## Advanced Exercise: Implementing Policy Iteration

In this advanced exercise, students will implement the policy iteration algorithm to find the optimal policy for the grid world MDP.

```python
# Policy Iteration Algorithm
def policy_iteration(grid_size, gamma, num_iterations):
    # Initialize a random policy
    policy = np.random.choice(actions, (grid_size, grid_size))
    value_function = np.zeros((grid_size, grid_size))

    for _ in range(num_iterations):
        # Policy Evaluation
        while True:
            new_value_function = np.copy(value_function)
            delta = 0
            for i in range(grid_size):
                for j in range(grid_size):
                    state = (i, j)
                    action = policy[i, j]
                    new_value_function[i, j] = calculate_expected_value(
                        state, action, grid_size, gamma, value_function)
                    delta = max(delta, abs(new_value_function[i, j] - value_function[i, j]))
            value_function = new_value_function
            if delta < 1e-3:
                break

        # Policy Improvement
        policy_stable = True
        for i in range(grid_size):
            for j in range(grid_size):
                state = (i, j)
                old_action = policy[i, j]
                action_values = {action: calculate_expected_value(state, action, grid_size, gamma, value_function)
                                 for action in actions}
                policy[i, j] = max(action_values, key=action_values.get)
                if old_action != policy[i, j]:
                    policy_stable = True

        if policy_stable:
            break
            
    return policy, value_function

optimal_policy, optimal_value_function = policy_iteration(grid_size, 0.9, 1000)

print("Optimal Policy:")
print(optimal_policy)
print("\nOptimal Value Function:")
print(optimal_value_function)
```

### Explanation:

This exercise demonstrates the policy iteration algorithm, which alternates between policy evaluation (estimating the value of a policy) and policy improvement (using the estimated values to find a better policy). This cycle continues until the policy becomes stable (i.e., it no longer changes between iterations), resulting in the optimal policy and its corresponding value function.

## Advanced Exercise: Implementing Upper Confidence Bound (UCB)

In this advanced exercise, we will implement the Upper Confidence Bound (UCB) algorithm for the one-armed bandit problem. This algorithm balances exploration and exploitation by considering the uncertainty in the action-value estimates.

```python
import math

# Define UCB function
def ucb(q_value, n, t, c):
    if n == 0:
        return float('inf')
    return q_value + c * math.sqrt(math.log(t) / n)

# Initialize variables
q_value = 0
n = 0
c = 2
total_rewards = []

# Simulate 1000 pulls with UCB
num_pulls = 1000
for t in range(1, num_pulls + 1):
    if np.random.random() < win_probability:
        reward = 1
    else:
        reward = 0
    q_value = (q_value * n + reward) / (n + 1)
    n += 1
    total_rewards.append(reward)

    ucb_value = ucb(q_value, n, t, c)
    print(f"Pull {t}: Reward = {reward}, UCB Value = {ucb_value}")

print(f"\nTotal rewards after {num_pulls} pulls: {sum(total_rewards)}")
```

### Explanation:

The UCB algorithm calculates an upper confidence bound for each action based on the estimated action-value (`q_value`) and the number of times the action has been taken (`n`). The algorithm balances exploration and exploitation by choosing the action with the highest upper confidence bound. We used a confidence parameter `c` of 2 in this implementation.

## Advanced Exercise: Comparing Epsilon-Greedy and UCB

In this advanced exercise, we will compare the performance of the Epsilon-Greedy strategy and the Upper Confidence Bound (UCB) algorithm in a multi-armed bandit setting.

```python
# Define epsilon-greedy function for multiple arms
def epsilon_greedy_multi(epsilon, n, t, q_values):
    if np.random.random() < epsilon:
        return np.random.choice(len(q_values))
    else:
        return np.argmax(q_values)

# Define UCB function for multiple arms
def ucb_multi(q_values, n_values, t, c):
    ucb_values = [ucb(q, n, t, c) for q, n in zip(q_values, n_values)]
    return np.argmax(ucb_values)

num_pulls = 1000
epsilon = 0.1
c = 2

# Initialize variables for epsilon-greedy
q_values_eg = np.zeros(len(win_probabilities))
n_values_eg = np.zeros(len(win_probabilities))
rewards_eg = []

# Initialize variables for UCB
q_values_ucb = np.zeros(len(win_probabilities))
n_values_ucb = np.zeros(len(win_probabilities))
rewards_ucb = []

# Simulate pulls
for t in range(1, num_pulls + 1):
    # Epsilon-Greedy
    arm_eg = epsilon_greedy_multi(epsilon, n_values_eg, t, q_values_eg)
    reward_eg = pull_bandit(arm_eg)
    n_values_eg[arm_eg] += 1
    q_values_eg[arm_eg] = (q_values_eg[arm_eg] * (n_values_eg[arm_eg] - 1) + reward_eg) / n_values_eg[arm_eg]
    rewards_eg.append(reward_eg)

    # UCB
    arm_ucb = ucb_multi(q_values_ucb, n_values_ucb, t, c)
    reward_ucb = pull_bandit(arm_ucb)
    n_values_ucb[arm_ucb] += 1
    q_values_ucb[arm_ucb] = (q_values_ucb[arm_ucb] * (n_values_ucb[arm_ucb] - 1) + reward_ucb) / n_values_ucb[arm_ucb]
    rewards_ucb.append(reward_ucb)

print(f"Total rewards (Epsilon-Greedy): {sum(rewards_eg)}")
print(f"Total rewards (UCB): {sum(rewards_ucb)}")
```

## Advanced Exercise: Implementing Thompson Sampling

In this advanced exercise, we will implement Thompson Sampling, a probabilistic method for balancing exploration and exploitation in multi-armed bandit problems.

```python
# Initialize variables for Thompson Sampling
successes = np.zeros(len(win_probabilities))
failures = np.zeros(len(win_probabilities))
total_rewards = []

# Define a function for Thompson Sampling
def thompson_sampling(successes, failures):
    sampled_probs = [np.random.beta(s+1, f+1) for s, f in zip(successes, failures)]
    return np.argmax(sampled_probs)

# Simulate pulls using Thompson Sampling
num_pulls = 1000
for _ in range(num_pulls):
    arm = thompson_sampling(successes, failures)
    reward = pull_bandit(arm)
    if reward == 1:
        successes[arm] += 1
    else:
        failures[arm] += 1
    total_rewards.append(reward)

print(f"Total rewards after {num_pulls} pulls: {sum(total_rewards)}")
```

### Explanation:

Thompson Sampling uses Bayesian inference to sample from the posterior distribution of the success probabilities for each bandit arm. The arm with the highest sampled probability is selected, balancing exploration and exploitation based on the observed successes and failures.

# Topic 7: Q LEARNING AND TEMPORAL DIFFERENCE LEARNING


## Exercise: Q-Learning Update Rule

In this exercise, we will implement the Q-learning update rule, a fundamental component of the Q-learning algorithm.

```python
# Define Q-learning update rule
def q_learning_update(q_values, state, action, reward, next_state, alpha, gamma):
    best_next_action = np.argmax(q_values[next_state])
    td_target = reward + gamma * q_values[next_state][best_next_action]
    q_values[state][action] += alpha * (td_target - q_values[state][action])

# Define a simple MDP for testing
states = [0, 1, 2]
actions = [0, 1]
q_values = np.zeros((len(states), len(actions)))
alpha = 0.1
gamma = 0.9

# Update Q-values for a sample transition
state = 0
action = 1
reward = 10
next_state = 2
q_learning_update(q_values, state, action, reward, next_state, alpha, gamma)
print("Updated Q-values:")
print(q_values)
```

### Explanation:

We implemented the Q-learning update rule, which updates the Q-values based on the reward received and the estimated future rewards. This rule is crucial for learning the optimal policy in Q-learning.

## Exercise: Q-Learning on Grid World

In this exercise, we will implement the Q-learning algorithm on the 4x4 grid world environment to learn the optimal policy.

```python
# Define grid world size and actions
grid_size = 4
actions = ['Up', 'Down', 'Left', 'Right']
action_map = {0: 'Up', 1: 'Down', 2: 'Left', 3: 'Right'}

# Define the transition and reward structure
def transition(state, action):
    i, j = state
    if action == 'Up':
        next_state = (max(i-1, 0), j)
    elif action == 'Down':
        next_state = (min(i+1, grid_size-1), j)
    elif action == 'Left':
        next_state = (i, max(j-1, 0))
    elif action == 'Right':
        next_state = (i, min(j+1, grid_size-1))
    else:
        next_state = state
        
    reward = -1
    if next_state == (grid_size-1, grid_size-1):
        reward = 10
    return next_state, reward

# Define Q-learning parameters
alpha = 0.1
gamma = 0.9
epsilon = 0.1
num_episodes = 500

# Initialize Q-values
q_values = np.zeros((grid_size, grid_size, len(actions)))

# Epsilon-greedy policy
def epsilon_greedy_policy(state, q_values, epsilon):
    if np.random.rand() < epsilon:
        return np.random.choice(len(actions))
    return np.argmax(q_values[state])

# Q-learning algorithm
for _ in range(num_episodes):
    state = (0, 0)
    while state != (grid_size-1, grid_size-1):
        action = epsilon_greedy_policy(state, q_values, epsilon)
        next_state, reward = transition(state, actions[action])
        q_learning_update(q_values, state, action, reward, next_state, alpha, gamma)
        state = next_state

# Print optimal policy
optimal_policy = np.array([[action_map[np.argmax(q_values[i, j])] for j in range(grid_size)] for i in range(grid_size)])
print("Optimal Policy:")
print(optimal_policy)
```

### Explanation:

We implemented the full Q-learning algorithm on a 4x4 grid world. The agent starts from the top-left corner and aims to reach the bottom-right corner. The Q-values are updated based on the Q-learning update rule. The optimal policy is derived from the Q-values after running multiple episodes.

## Advanced Exercise: Implementing SARSA

In this advanced exercise, students will implement the SARSA (State-Action-Reward-State-Action) algorithm on the same grid world environment to compare the results with Q-learning.

```python
# Initialize Q-values for SARSA
q_values_sarsa = np.zeros((grid_size, grid_size, len(actions)))

# SARSA update rule
def sarsa_update(q_values, state, action, reward, next_state, next_action, alpha, gamma):
    td_target = reward + gamma * q_values[next_state][next_action]
    q_values[state][action] += alpha * (td_target - q_values[state][action])

# SARSA algorithm
for _ in range(num_episodes):
    state = (0, 0)
    action = epsilon_greedy_policy(state, q_values_sarsa, epsilon)
    while state != (grid_size-1, grid_size-1):
        next_state, reward = transition(state, actions[action])
        next_action = epsilon_greedy_policy(next_state, q_values_sarsa, epsilon)
        sarsa_update(q_values_sarsa, state, action, reward, next_state, next_action, alpha, gamma)
        state, action = next_state, next_action

# Print optimal policy for SARSA
optimal_policy_sarsa = np.array([[action_map[np.argmax(q_values_sarsa[i, j])] for j in range(grid_size)] for i in range(grid_size)])
print("Optimal Policy (SARSA):")
print(optimal_policy_sarsa)
```

### Explanation:

We implemented the SARSA algorithm on the same 4x4 grid world environment. Unlike Q-learning, SARSA updates its Q-values based on the action taken (`a'`) in the next state (`s'`). We compared the optimal policies derived from both Q-learning and SARSA to see the differences in the learned behaviors.

## Advanced Exercise: Implementing Deep Q-Learning Algorithm

In this advanced exercise, we will implement the full Deep Q-Learning algorithm using the defined Q-network and experience replay.

```python
from collections import deque
import random

# Hyperparameters
learning_rate = 0.001
gamma = 0.95
epsilon = 1.0
epsilon_min = 0.01
epsilon_decay = 0.995
batch_size = 32
memory = deque(maxlen=2000)

# Experience replay memory
def remember(state, action, reward, next_state, done):
    memory.append((state, action, reward, next_state, done))

# Get action using epsilon-greedy policy
def act(model, state):
    if np.random.rand() <= epsilon:
        return random.choice(range(num_actions))
    q_values = model.predict(state)
    return np.argmax(q_values[0])

# Replay function
def replay(batch_size):
    global epsilon
    minibatch = random.sample(memory, batch_size)
    for state, action, reward, next_state, done in minibatch:
        target = reward
        if not done:
            target += gamma * np.amax(model.predict(next_state)[0])
        target_f = model.predict(state)
        target_f[0][action] = target
        model.fit(state, target_f, epochs=1, verbose=0)
    if epsilon > epsilon_min:
        epsilon *= epsilon_decay

# Training the Deep Q-Network
for e in range(num_episodes):
    state = (0, 0)
    state_input = state_to_input(state, grid_size)
    for time in range(500):
        action = act(model, state_input)
        next_state, reward = transition(state, actions[action])
        next_state_input = state_to_input(next_state, grid_size)
        done = next_state == (grid_size-1, grid_size-1)
        remember(state_input, action, reward, next_state_input, done)
        state = next_state
        state_input = next_state_input
        if done:
            print(f"Episode: {e}/{num_episodes}, time: {time}, reward: {reward}")
            break
    if len(memory) > batch_size:
        replay(batch_size)

# Testing the trained model
optimal_policy_nn = np.array([[action_map[np.argmax(model.predict(state_to_input((i, j), grid_size))[0])] for j in range(grid_size)] for i in range(grid_size)])
print("Optimal Policy (NN):")
print(optimal_policy_nn)
```
